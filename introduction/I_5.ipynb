# Curse of Dimensionality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
from numpy.random import uniform
import pylab

def sample(d=2, N=1000):
    return [[uniform(0., 1.) for i in range(d)] for _ in range(N)]

def corner_count(points):
    return np.mean([any([(d < .01 or d > .99) for d in p]) for p in points])

def go(Ds=range(1,200)):
    plt.plot(Ds, [corner_count(sample(d)) for d in Ds])
```

```python
data=sample(d=2, N=1000)
data_1=[any([(d < .01 or d > .99) for d in p]) for p in data]
frames = [pd.DataFrame(data, columns=list(["data1","data2"])), pd.DataFrame(list(map(int,data_1)), columns=list(["labes"]))]
df=pd.concat(frames, axis=1)
sns.pairplot(df, hue="labes",vars=["data1","data2"], diag_kind="kde")
plt.show()
```

```python
data=sample(d=5, N=1000)
data_1=[any([(d < .01 or d > .99) for d in p]) for p in data]
frames = [pd.DataFrame(data, columns=list(["data1","data2","data3","data4","data5"])), pd.DataFrame(list(map(int,data_1)), columns=list(["labes"]))]
df=pd.concat(frames, axis=1)
sns.pairplot(df, hue="labes",vars=["data1","data2","data3","data4","data5"], diag_kind="kde")
plt.show()
```

```python
go(Ds=range(1,200))
plt.show()
```
