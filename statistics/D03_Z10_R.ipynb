# Rozkład logarytmicznie normalny

Rozkład logarytmicznie normalny ma gęstość

$$
LN(x;\mu,\sigma^2) = 
\left\{ \begin{array}{ll}
{\displaystyle {\frac {1}{\sigma {\sqrt {2\pi }}}}\ e^{-{\frac {\left(\ln x-\mu \right)^{2}}{2\sigma ^{2}}}}} & \textrm{gdy $x \geq 0$}\\
0 & \textrm{gdy $x < 0$}
\end{array} \right. .
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
t = np.arange(0, 10, 0.05)
(mu, sigma)=(1,1)
lognorm=stats.lognorm(mu, sigma)
```

# Narysujmy gętość zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
plt.plot(t, lognorm.pdf(t), '--', ms=8, label='frozen pdf')
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, lognorm.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = lognorm.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, lognorm.pdf(t), 'k-', lw=2, label='a=-1, b=1')
ax.legend()
plt.show()
```

# Narysuj gęstości rozkładu jednostajnego z różnymi parametrami

```python
sigma_values = [0.5, 1.0, 2.0]
mu_values = [1, 1, 2]
fig, ax = plt.subplots(1, 1)
for mu, sigma in zip(mu_values, sigma_values):
    dist = stats.lognorm(mu, sigma)
    plt.plot(t, dist.pdf(t), ls="--", label=r'$\mu=%1.f,\ \sigma=%.1f$' % (mu, sigma))
plt.legend()
plt.show()
```

# Narysuj kilka histogramów różnych próbek z roskładu jednostajnego o tych samych parametrach.

```python
norm=stats.lognorm(1,1)
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = norm.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
