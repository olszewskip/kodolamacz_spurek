# Rozkład Weibulla
Rozkład Weibulla jest najczęściej stosowany do analizy przeżycia. 

Rozkład Weibulla  ma gęstość
$$
f_{X}(x;k,\lambda) = 
\left\{ \begin{array}{ll}
\frac{k}{\lambda} \left( \frac{k}{\lambda} \right)^{k-1} e^{-(x/\lambda)^k} & \textrm{gdy $x \geq 0$}\\
0 & \textrm{gdy $x < 0$}
\end{array} \right. ,
$$
gdzie $k>0$ jest parametrem kształtu oraz $\lambda> 0$ jest parametrem skali.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
t = np.arange(0, 5, 0.1)
lambdaVal = 1
k = 2
X=stats.weibull_min(k,lambdaVal)
```

# Narysujmy gętość zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
plt.plot(t, X.pdf(t), '--', ms=8, label='frozen pdf')
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, X.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = X.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, X.pdf(t), 'k-', lw=2, label='a=-1, b=1')
ax.legend()
plt.show()
```

# Narysuj gęstości rozkładu jednostajnego z różnymi parametrami

```python
lambda_values = [1.5, 3.0, 2.0]
k_values = [1, 1, 2]
fig, ax = plt.subplots(1, 1)
for k, lambdaVal in zip(k_values, lambda_values):
    dist = stats.weibull_min(k, lambdaVal)
    plt.plot(t, dist.pdf(t), ls="--", label=r'$k=%1.f, \lambda=%.1f$' % (k, lambdaVal))
plt.legend()
plt.show()
```

# Narysuj kilka histogramów różnych próbek z roskładu jednostajnego o tych samych parametrach.

```python
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = X.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
